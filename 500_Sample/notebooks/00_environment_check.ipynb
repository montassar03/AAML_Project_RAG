# Environment Check

This notebook is used to record the runtime environment for the project:
- Python executable
- Python version
- package versions
- GPU information
- other relevant setup details
- The work is on the lmu shared servers
- GPU-based experiments will need to be launched through Slurm.

```python
import sys
import torch
import transformers
import datasets
import faiss
import sentence_transformers
import pandas
import numpy
import importlib.metadata as md

print("Python executable:", sys.executable)
print("Python version:", sys.version)
print("torch:", torch.__version__)
print("transformers:", transformers.__version__)
print("datasets:", datasets.__version__)
print("llama-index:", md.version("llama-index"))
print("llama-index-core:", md.version("llama-index-core"))
print("faiss:", faiss.__version__)
print("sentence_transformers:", sentence_transformers.__version__)
print("numpy:", numpy.__version__)
print("pandas:", pandas.__version__)
print("CUDA available:", torch.cuda.is_available())
```

```python
import importlib

def check_version(pkg_name):
    try:
        pkg = importlib.import_module(pkg_name)
        version = getattr(pkg, "__version__", "no __version__ attribute")
        print(f"{pkg_name}: {version}")
    except ImportError:
        print(f"{pkg_name}: NOT INSTALLED")

# Core libraries for your project
packages = [
    "torch",
    "transformers",
    "datasets",
    "llama_index",
    "faiss",
    "sentence_transformers",
    "numpy",
    "pandas"
]

for p in packages:
    check_version(p)
```

```python
import importlib.metadata as md

packages = [
    "llama-index",
    "llama-index-core"
]

for pkg in packages:
    try:
        print(f"{pkg}: {md.version(pkg)}")
    except md.PackageNotFoundError:
        print(f"{pkg}: NOT INSTALLED")
```
